```python
Data = {'Year': [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010],
       'Unemployment_Rate':[ 9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
       }
```

```python
plt.plot(unemployment['Year'],
        unemployment['Unemployment_Rate'],
        color='blue',
        marker="^",
        ls = '--')
plt.show()
```

```python
x = np.arange(6)
fig=plt.figure()
fig.show()
ax=fig.add_subplot(111)

ax.plot(x, x, c='b', marker="^", ls='--', label='L1', fillstyle='none')
ax.plot(x, x+1, c='g', marker=(8, 2, 0), ls='--', label='L2')
ax.plot(x, (x+1)**2, c='k', ls='-', label='L3')
ax.plot(x, (x-1)**2, c='r', marker="v", ls='-', label='L4')
ax.plot(x, x**2-1, c='m', marker="o", ls='--', label='L5', fillstyle='none')
ax.plot(x, x-1, c='k', marker="+", ls=':', label='L6')

plt.legend(loc=2)
plt.draw()
```

```python
fig = plt.subplots(figsize=(10, 10))
sns.heatmap(diamonds.corr(), annot=True, fmt=".2f")
```

What can we understand from the visual?
- From this heatmap of correlation, we can understand that the darker parts (negative numbers) indicate a negative correlation between the two parameters being compared. For example, table and depth have a very high - correlation which indicates that as one paramter increases, the other reduces.

- Anything close to 0 means no great correlation between the parameters is seen, such as with parameters 'Depth' and 'Table' with most other parameters. 

- Price and Carat, have a good positive correlation, which implies that as carats increase, the price of the diamond increases.

- X is the length, Y is the width and Z is the depth of a diamond. Price and 'Y', 'X', 'Z', have a good positive correlation, which implies that as length/breadth/depth of the diamond increases, the price of the diamond increases.

- X is the length, Y is the width and Z is the depth of a diamond. Carat and 'Y', 'X', 'Z', have a good positive correlation, which implies that as length/breadth/depth of the diamond increases, the carat value of the diamond increases.
